# 데이터 불러오기

```python
import pandas as pd

# data read
data1 = pd.read_csv('/Users/yul/Desktop/LP Player/3. LP Player_데이터 및 모델 세이브 파일/LPOINT_BIG_COMP_01_DEMO.csv')
data2 = pd.read_csv('/Users/yul/Desktop/LP Player/3. LP Player_데이터 및 모델 세이브 파일/LPOINT_BIG_COMP_02_PDDE.csv')
data3 = pd.read_csv('/Users/yul/Desktop/LP Player/3. LP Player_데이터 및 모델 세이브 파일/LPOINT_BIG_COMP_04_PD_CLAC.csv')
```

```python
data1 = pd.DataFrame(data1)
data2 = pd.DataFrame(data2)
data3 = pd.DataFrame(data3)
```

## DEMO

```python
# 30~40대 여성, 서울/경기 지역 고객만
data1 = data1[data1['ma_fem_dv']=='여성']
condition_1 = (data1['ages'] == '30대') | (data1['ages'] == '40대')
condition_2 = (data1['zon_hlv'] == 'Z10') | (data1['zon_hlv'] == 'Z17')
data1 = data1[condition_1 & condition_2]
data1
```

```python
data1.info()
```

```python
data1.nunique()
```

## PDDE

```python
# 월, 요일 추가
data2['de_dt'] = data2['de_dt'].astype('str')
data2['de_dt'] = pd.to_datetime(data2['de_dt'])
data2['de_dt_month'] = data2['de_dt'].dt.month
data2['de_day'] = data2['de_dt'].dt.dayofweek
data2
```

```python
# 데이터 타입 변환
data2['buy_am'] = data2['buy_am'].astype('int64')
data2['chnl_dv'] = data2['chnl_dv'].astype('str')
```

```python
data2.info()
```

```python
data2.nunique()
```

# 데이터 병합

```python
data = pd.merge(data1, data2, on='cust', how='inner')
data = pd.merge(data, data3, on='pd_c', how='inner')
data
```
