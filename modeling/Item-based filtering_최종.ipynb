```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mean_squared_error

# 데이터 불러오기
data = pd.read_pickle("/Users/yul/Desktop/LP Player/3. LP Player_데이터 및 모델 세이브 파일/Item-based_filtering_data_pickle_최종")
np.random.seed(19670324)  # seed 값 지정
```

# pivot_table 만들기

```python
ratings = data.pivot_table(index='pd_nm', columns='cust', values='buy_ct')
ratings.fillna(0, inplace=True)  # na값 0으로 대체
ratings
```

# 훈련 데이터와 테스트 데이터 분리

```python
ratings_train, ratings_test = train_test_split(ratings, test_size=0.25, random_state=42)
ratings_train.shape, ratings_test.shape
```

# 코사인 유사도

```python
item_similarity = cosine_similarity(ratings_train)   # ratings_train의 아이템 유사도
item_similarity
```

# 예측값 계산

```python
# 예측값 계산
item_pred = item_similarity.dot(ratings_train) / np.array([np.abs(item_similarity).sum(axis=1)]).T
item_pred_matrix = pd.DataFrame(data=item_pred, index=ratings_train.index, columns=ratings_train.columns)
item_pred_matrix.head(10)
```

```python
item_pred_matrix.fillna(0, inplace=True)
```

# 아이템 기반 상품 추천

```python
# item_similarity(아이템 유사도)는 array형태로 열 이름이 없기 때문에 DataFrame으로 바꿔주고 유사도 높은 고객을 찾아냄
item_similarity_df = pd.DataFrame(
    data=item_similarity,
    index=ratings_train.index,
    columns=ratings_train.index)

# 가장 유사도가 높은 top10
def recommand_pd(pd_nm):
    return item_similarity_df[pd_nm].sort_values(ascending=False)[:11]
```

```python
recommand_pd('젤리')
```

# 모델 평가 지표 : RMSE

```python
# RMSE는 DataFrame으로 계산 불가 -> array로 바꿔줌
ratings_train_np = np.array(ratings_train)
ratings_test_np = np.array(ratings_test)

# RMSE 함수
def get_rmse(pred, actual):
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred, actual)**0.5 # RMSE
```

```python
print("train:", get_rmse(item_pred, ratings_train_np)) # 훈련에 대한 평가지표
print("test:", get_rmse(item_pred, ratings_test_np)) # 테스트에 대한 평가지표
```

# 군집 분석을 이용한 알고리즘 개선
- k-NN을 이용하여 가장 비슷한 몇 개의 아이템끼리만 비교하도록 개선

```python
k=5
neigh = NearestNeighbors(n_neighbors=k, metric="cosine")
neigh.fit(ratings_train)
top_k_distances, top_k_items = neigh.kneighbors(ratings_train, return_distance=True)
top_k_similarity = 1 - top_k_distances
top_k_similarity
```

# 군집 분석 후 RMSE 평가

```python
item_pred_k = np.zeros(ratings_train_np.shape)

for i in range(ratings_train_np.shape[0]):
    item_pred_k[i, :] = top_k_similarity[i].T.dot(ratings_train_np[top_k_items][i]) / np.array([np.abs(top_k_similarity[i].T).sum(axis=0)]).T
```

```python
item_pred_k = np.nan_to_num(item_pred_k)
```

```python
print("train:",get_rmse(item_pred_k, ratings_train_np))
print("test:",get_rmse(item_pred_k, ratings_test_np))
```
